```python
# the transactions are inside the interval bucket end and bucket start dates?
agg = db.listingsAndReviews_new.aggregate([
    {
        '$addFields': {
            'oldest_transaction': { '$min': '$transactions.transactions.date' }, 
            'most_recent_transaction': { '$max': '$transactions.transactions.date' } 
        }
    },
    {
        '$project': {
            '_id': 1,  # Include the document ID
            'transactions.bucket_end_date': 1,
            'transactions.bucket_start_date': 1,
            'oldest_transaction': 1,  # Include the oldest transaction datetime
            'most_recent_transaction': 1  # Include the most recent transaction datetime
        }
    }
])

for doc in agg:
    id = doc.get('_id')
    bucket_start_date = doc['transactions']['bucket_start_date']
    bucket_end_date = doc['transactions']['bucket_end_date']
    oldest_transaction = doc.get('oldest_transaction')
    most_recent_transaction = doc.get('most_recent_transaction')

    if (oldest_transaction >= bucket_start_date) and (most_recent_transaction <= bucket_end_date):
        inside_interval = True
    else:
        inside_interval = False
        print(f"The listing with the id {id} does not have all transactions inside the interval!")

if inside_interval:
    print("All transactions are inside the interval")
```

```python
# are there duplicated date in the transactions in the transactions field from transactions collection?
pipeline = [
    {"$project": {"transactions":1}},
    # Unwind the transactions array to inspect each transaction
    {"$unwind": "$transactions.transactions"},
    # Group by listing _id AND date to find duplicates within each listing
    {"$group": {
        "_id": {
            "listing_id": "$_id",
            "date": "$transactions.transactions.date"
        },
        "count": {"$sum": 1}
    }},
    # Only keep groups where the same date appears >1 time in a listing
    {"$match": {"count": {"$gt": 1}}},
    # Reshape the output for clarity
    {"$project": {
        "listing_id": "$_id.listing_id",
        "date": "$_id.date",
        "count": 1,
        "_id": 0
    }},
    # Sort by listing_id and count
    {"$sort": {"count": -1}}
]

for doc in db.listingsAndReviews_new.aggregate(pipeline):
    pprint(doc)
```

#### Schema transformations and changes

- **New collections:** 
    - *listings* -> We include what should be important to know when we search for a specific listing, as a mostly reading-wise document.
    - *reviews* -> Also, we add the hosts_id as a field in the reviews, because what if the host changes? we want to keep that respective host of that review, in that moment in time.?
    - *hosts* -> A separate collection for the hosts help to keep track in the database of every host that has ever registered and had a listing, and consequently have had a review, even if the listing is changed or its host changed. Could improve performance, since we could just search for a specific host or the reviews or listings related to one only. And when we look for a specific listing we are not interested in the full information of that host. 
    - *transactions* -> There is no need to retrieve the transactions information alongside witth the listings information. Also, this could constitute a very long array of transactions documents. For this reasons, the best practice is to separate the transactions information to another collection.
- **Treat missing values**
    - Such as '.', '*' -> Change to an empty string (they will not be set to null, since the lack of value could be also viewed as an important information in the near future).
- **Correct data types**
    - minimum_nights and maximum_nights -> to integers
    - rounding the price in the transactions to decimal
- **Attribute Patterns**
    - review scores fields -> Attribute pattern is applied as these fields are not unanimous throughout the documents. This way, it becomes more versatile as to adding new review score information. Also, it is better considering the queries wanted refered in the advanced questions.

- **Embedded document for hosts**
    - hosts-related fields -> Although there are two fields that are not consistent throughout all documents, we should keep each field as new subfield whithin the hosts, as we do not need to comupte any metric between those but rather to display it in a cleaner way for the user.

    *Key note*:
    Review overall scores and hosts information is kept in the listings as it is an information that should be returned together for when we search for a specific listing
- **Duplicated fields:**
    - last_scraped and calendar_last_scraped
    - reviews and reviews copys (1 to 4)
    - host_listings_count and host_total_listings_count
- **Non-important fields**:
    - first_review and last_review -> since, if necessary, it can be accessed through a query in the reviews collection, by sorting by the date.´
- **Treating the transactions**:
    - Removing the transactions' prices that dont meet the minimum price criteria for that listing. -> We notice that there are very low prices that do not seem to match a listings price (such as 3 monetary units). For this reason, we calculate the price that should supposedly be the minimum accepted.
    - Why do we not remove the very old transactions in the DataBase? Although they are whithin a range between the bucket end and bucket start date, we searched for the creation of the Airbnb, which started in 2007. However, this is information that was not given with the project. Therefore, we consider that we should not take into account the external information found, but rather the oficial bucket end and start dates that were presented in the DB.

```python
# the fields minimum_nights and maximum_nights should be integers
db.listingsAndReviews_new.update_many(
    {}, 
    [{'$set': {'minimum_nights': { '$toInt': '$minimum_nights' },
                'maximum_nights': { '$toInt': '$maximum_nights' }}
    }]
)

for doc in db.listings.find({}, {'minimum_nights': 1, 'maximum_nights': 1}):
    if not isinstance(doc['minimum_nights'], int):
        print(f"minimum_nights is not an integer for document with _id: {doc['_id']}")
    if not isinstance(doc['maximum_nights'], int):
        print(f"maximum_nights is not an integer for document with _id: {doc['_id']}")
```

```python
# Convert the price field from a string to an integer and round it to 2 decimal places
db.listingsAndReviews_new.update_many(
    {},  # Update all documents
    [{'$set': {'transactions.transactions': {
                    '$map': {'input': '$transactions.transactions',
                             'as': 'transaction',
                             'in': {'$mergeObjects': ['$$transaction',  # Keep all existing fields in the transaction
                                {'price': {'$round': [{'$toDouble': '$$transaction.price'}, 2]}}]  # Convert price to double and round
                                    }
                            }}
            }
    }]
)

doc = db.listingsAndReviews_new.find_one({}, {'transactions': 1, '_id': 0})
print(f'The price from transactions changed to {type(doc['transactions']['transactions'][0]['price'])} datatype and was rounded to 2 decimal places')
```

> Treat missing values

```python
# Change fields with values '.' or '*' to an empty string ("")
db.listingsAndReviews_new.update_many(
    {},  # Update all documents
    [{'$set': {'newRoot': {'$arrayToObject': {
                        '$map': {'input': { '$objectToArray': '$$ROOT' },
                                'as': 'field',
                                'in': {'k': '$$field.k',
                                       'v': {'$cond': {'if': {'$or': [
                                                                { '$eq': ['$$field.v', '.'] },
                                                                { '$eq': ['$$field.v', '*'] }]},
                                                        'then': "",
                                                        'else': '$$field.v'}}}
                                }}
                        }}},
    {'$replaceRoot': {'newRoot': '$newRoot'}}
    ])
```

```python
# Trim strings for fields with strings
db.listingsAndReviews_new.update_many(
    {},  # Update all documents
    [{'$replaceRoot': {'newRoot': {'$arrayToObject': {
                        '$map': {
                            'input': { '$objectToArray': '$$ROOT' },  # Convert the document to an array of key-value pairs
                            'as': 'field',
                            'in': {'k': '$$field.k',
                                   'v': {'$cond': {
                                        'if': { '$eq': [{ '$type': '$$field.v' }, 'string'] },  # Check if the value is a string
                                        'then': { '$trim': { 'input': '$$field.v' } },  # Trim the string
                                        'else': '$$field.v'}}}
                        }}}}
    }])
```

```python
db.reviews.create_index("listing_id")


def process_batch(batch):
    bulk_operations = []
    for listing in batch:
        listing_id = listing['_id'] 
        host_id = listing.get('host_id') 
        reviews = listing.get('reviews', [])

        if reviews:  #ensure reviews exist
            reviews_sorted = sorted(reviews, key=lambda x: x.get('date', datetime.min), reverse=True)
            db.listings.update_one(
                {"_id": listing_id},
                {"$set": {"reviews": reviews_sorted[:10]}})  #left just the 10 most recent reviews

            for review in reviews_sorted:
                review['listing_id'] = listing_id
                review['host_id'] = host_id #add the host_id
                bulk_operations.append(UpdateOne(
                    {"_id": review.get("_id")},
                    {"$set": review},
                    upsert=True)) #update or insert the review
        
    if bulk_operations:
        db.reviews.bulk_write(bulk_operations)

def migrate(): #function to do it in parallel
    batch_size = 1111
    total_documents = db.listings.count_documents({})
    num_threads = 4  #number of threads for parallel processing

    with ThreadPoolExecutor(max_workers=num_threads) as executor:
        futures = []
        for i in range(0, total_documents, batch_size):
            batch = list(db.listings.find().skip(i).limit(batch_size))
            futures.append(executor.submit(process_batch, batch))
        for future in futures:
            future.result()
    
migrate()

print("Reviews have been successfully migrated to the new collection, and only the 10 most recent reviews are retained in the original collection.")
```

```python
db.transactions.create_index("listing_id")

def process_batch(batch):
    bulk_operations = []
    for listing in batch:
        listing_id = listing['_id']  # Get the _id of the listing
        transactions = listing.get('transactions')  # Get the transactions subdocument

        if transactions:
            transactions['listing_id'] = listing_id
            bulk_operations.append(UpdateOne(
                {"listing_id": listing_id},  # Filter by listing_id
                {"$set": transactions},  # Update or insert the transactions subdocument
                upsert=True))
            db.listings.update_one(
                {"_id": listing_id},
                {"$unset": {"transactions": ""}})
    
    if bulk_operations:
        db.transactions.bulk_write(bulk_operations)

migrate()

print("Transactions have been successfully migrated to the new collection and removed from the original collection.")
```

```python
result = db.listings.aggregate([
    {'$match': {'accommodates': {'$gt': 4}, 
                'hosts.is_superhost': True
                }},
    {'$group': {'_id': '$hosts.id',
                'qualified_listings_count': {'$sum': 1}, #because it already consists of the listings that can accommodate more than 4 guests
                'host_data': {'$first': '$hosts'}, #keep the host data for later projection
                }},
    {'$match': {'qualified_listings_count': {'$gte': 2}
                }},
    {'$limit': 3},
    {'$project': {
        'id': '$host_data.id',
        'total_listings': '$host_data.listings_count',
        'qualified_listings': '$qualified_listings_count',
    }}
])

for h in result:
    print(f"Host with id {h['id']} selected. It is a superhost and has {h['total_listings']} listings, where {h['qualified_listings']} of these can accommodate more than 4 guests.")
```

### Question 3

The company considers investing into property to rent. Which bed type is most common in listings with a waterfront and a dishwasher in New York?

```python
#most common bed type in the listings collection
result = db.listings.aggregate([
    {'$match': {'amenities': {'$all': ['Waterfront', 'Dishwasher']}}},
    {'$group': {
        '_id': '$bed_type',
        'count': {'$sum': 1}
    }},
    {'$project': {
        '_id': 0,
        'bed_type': '$_id',
        'count': 1
    }},
    {'$sort': {'count': -1}},
    {'$limit': 3}
])

for doc in result:
    print(f"In listings with a waterfront and a dishwasher, the most common bed type is '{doc['bed_type']}', with a frequency of {doc['count']} times.")
```

### Question 5

To assess the security of different areas, what is the biggest and smallest (price-security deposit) difference per number of visitors at a property?

```python
biggest_diff_pipeline = [{'$project':{'price':1,
                 'security_deposit':1,
                 'accommodates':1
                 }},
{'$addFields':{'difference':
                   {'$divide':[{'$abs':{'$subtract':['$price', '$security_deposit']}}, '$accommodates']}
                   }},
                   {'$sort':{'difference':-1}},
                   {'$limit':1}]

smallest_diff_pipeline = [{'$match': {
        'security_deposit': {'$ne': None},
    }},
    {'$project':{'price':1,
                 'security_deposit':1,
                 'accommodates':1
                 }},
{'$addFields':{'difference':
                   {'$divide':[{'$abs':{'$subtract':['$price', '$security_deposit']}}, '$accommodates']}
                   }},
                   {'$sort':{'difference':1}},
                   {'$limit':1}]

biggest_diff = [doc.get('difference') for doc in db.listings.aggregate(biggest_diff_pipeline)]
smallest_diff = [doc.get('difference') for doc in db.listings.aggregate(smallest_diff_pipeline)]

print(f"The biggest difference is: {biggest_diff}")
print(f"The smallest difference is: {smallest_diff}")
```

### Question 6

Identify areas by whether they are typically used for short breaks, like weekend mini breaks, or whether they are more suitable for long trips. This information support targeted advertising of different customer types. It is not expected to change much over time so we won’t look to update it, we just require current view. What is the average duration of stay (in nights) per type of property per city (you can use the maximum_nights to measure length of stays)? For each property type return the city with the highest and lowest average value.

```python
db.listings.find_one({}, {'address.market', 'property_type', 'maximum_nights'})
```

```python
pipeline = [
    {"$group": {
            "_id": {
                "property_type": "$property_type",
                "market": "$address.market"
            },
            "avg_max_nights": {"$avg": "$maximum_nights"}
        }
    },
    {
        "$sort": {"avg_max_nights": -1}
    }
]

results = list(db.listings.aggregate(pipeline))

property_stats = {}
for result in results:
    prop_type = result["_id"]["property_type"]
    market = result["_id"]["market"]
    avg = result["avg_max_nights"]
    
    if prop_type not in property_stats:
        property_stats[prop_type] = {
            "highest": {"market": market, "avg": avg},
            "lowest": {"market": market, "avg": avg}
        }
    else:
        if avg > property_stats[prop_type]["highest"]["avg"]:
            property_stats[prop_type]["highest"] = {"market": market, "avg": avg}
        if avg < property_stats[prop_type]["lowest"]["avg"]:
            property_stats[prop_type]["lowest"] = {"market": market, "avg": avg}

for prop_type, stats in property_stats.items():
    print(f"\nProperty Type: {prop_type}")
    print(f"Market with highest average: {stats['highest']['market']} ({round(stats['highest']['avg'], 2)} nights)")
    print(f"Market with lowest average: {stats['lowest']['market']} ({round(stats['lowest']['avg'], 2)} nights)")
```
