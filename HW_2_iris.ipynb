```python
# the number of transactions match the transactions_count?

docs = db.listingsAndReviews_new.find({}, {'transactions':1})

for doc in docs:
    id = doc.get('_id')
    transactions = doc.get('transactions', [])
    if len(transactions['transactions']) == transactions['transaction_count']:
        match = True
    else:
        match = False
        print(f"The transactions count does not match! The respective listing id is {id}")
    
if match:
    print("All transactions count match the number of transactions")
```

```python
# the transactions are inside the interval bucket end and bucket start dates?
agg = db.listingsAndReviews_new.aggregate([
    {
        '$addFields': {
            'oldest_transaction': { '$min': '$transactions.transactions.date' }, 
            'most_recent_transaction': { '$max': '$transactions.transactions.date' } 
        }
    },
    {
        '$project': {
            '_id': 1,  # Include the document ID
            'transactions.bucket_end_date': 1,
            'transactions.bucket_start_date': 1,
            'oldest_transaction': 1,  # Include the oldest transaction datetime
            'most_recent_transaction': 1  # Include the most recent transaction datetime
        }
    }
])

for doc in agg:
    id = doc.get('_id')
    bucket_start_date = doc['transactions']['bucket_start_date']
    bucket_end_date = doc['transactions']['bucket_end_date']
    oldest_transaction = doc.get('oldest_transaction')
    most_recent_transaction = doc.get('most_recent_transaction')

    if (oldest_transaction >= bucket_start_date) and (most_recent_transaction <= bucket_end_date):
        inside_interval = True
    else:
        inside_interval = False
        print(f"The listing with the id {id} does not have all transactions inside the interval!")

if inside_interval:
    print("All transactions are inside the interval")
```

```python
# the fields minimum_nights and maximum_nights should be integers
db.listingsAndReviews_new.update_many(
    {}, 
    [{'$set': {'minimum_nights': { '$toInt': '$minimum_nights' },
                'maximum_nights': { '$toInt': '$maximum_nights' }}
    }]
)

for doc in db.listings.find({}, {'minimum_nights': 1, 'maximum_nights': 1}):
    if not isinstance(doc['minimum_nights'], int):
        print(f"minimum_nights is not an integer for document with _id: {doc['_id']}")
    if not isinstance(doc['maximum_nights'], int):
        print(f"maximum_nights is not an integer for document with _id: {doc['_id']}")
```

```python
# Convert the price field from a string to an integer and round it to 2 decimal places
db.listingsAndReviews_new.update_many(
    {},  # Update all documents
    [{'$set': {'transactions.transactions': {
                    '$map': {'input': '$transactions.transactions',
                             'as': 'transaction',
                             'in': {'$mergeObjects': ['$$transaction',  # Keep all existing fields in the transaction
                                {'price': {'$round': [{'$toDouble': '$$transaction.price'}, 2]}}]  # Convert price to double and round
                                    }
                            }}
            }
    }]
)

doc = db.listingsAndReviews_new.find_one({}, {'transactions': 1, '_id': 0})
print(f'The price from transactions changed to {type(doc['transactions']['transactions'][0]['price'])} datatype and was rounded to 2 decimal places')
```

> Treat missing values

```python
# Remove fields with values '', '.', or '*'
db.listingsAndReviews_new.update_many(
    {},  # Update all documents
    [
        {
            '$replaceRoot': {
                'newRoot': {
                    '$arrayToObject': {
                        '$filter': {
                            'input': { '$objectToArray': '$$ROOT' },  # Convert the document to an array of key-value pairs
                            'as': 'field',
                            'cond': {
                                '$and': [
                                    { '$ne': ['$$field.v', ''] },  # Exclude empty string
                                    { '$ne': ['$$field.v', '.'] },  # Exclude '.'
                                    { '$ne': ['$$field.v', '*'] }   # Exclude '*'
                                ]
                            }
                        }
                    }
                }
            }
        }
    ]
)
```

```python
# Trim strings for fields with strings
db.listingsAndReviews_new.update_many(
    {},  # Update all documents
    [
        {
            '$replaceRoot': {
                'newRoot': {
                    '$arrayToObject': {
                        '$map': {
                            'input': { '$objectToArray': '$$ROOT' },  # Convert the document to an array of key-value pairs
                            'as': 'field',
                            'in': {
                                'k': '$$field.k',  # Keep the key as is
                                'v': {
                                    '$cond': {
                                        'if': { '$eq': [{ '$type': '$$field.v' }, 'string'] },  # Check if the value is a string
                                        'then': { '$trim': { 'input': '$$field.v' } },  # Trim the string
                                        'else': '$$field.v'  # Keep the value as is for non-strings
                                    }
                                }
                            }
                        }
                    }
                }
            }
        }
    ]
)
```

```python
db.reviews.create_index("listing_id")


def process_batch(batch):
    bulk_operations = []
    for listing in batch:
        listing_id = listing['_id'] 
        host_id = listing.get('host_id') 
        reviews = listing.get('reviews', [])

        if reviews:  #ensure reviews exist
            reviews_sorted = sorted(reviews, key=lambda x: x.get('date', datetime.min), reverse=True)
            db.listings.update_one(
                {"_id": listing_id},
                {"$set": {"reviews": reviews_sorted[:10]}})  #left just the 10 most recent reviews

            for review in reviews_sorted:
                review['listing_id'] = listing_id
                review['host_id'] = host_id #add the host_id
                bulk_operations.append(UpdateOne(
                    {"_id": review.get("_id")},
                    {"$set": review},
                    upsert=True)) #update or insert the review
        
    if bulk_operations:
        db.reviews.bulk_write(bulk_operations)

def migrate(): #function to do it in parallel
    batch_size = 1111
    total_documents = db.listings.count_documents({})
    num_threads = 4  #number of threads for parallel processing

    with ThreadPoolExecutor(max_workers=num_threads) as executor:
        futures = []
        for i in range(0, total_documents, batch_size):
            batch = list(db.listings.find().skip(i).limit(batch_size))
            futures.append(executor.submit(process_batch, batch))
        for future in futures:
            future.result()
    
migrate()

print("Reviews have been successfully migrated to the new collection, and only the 10 most recent reviews are retained in the original collection.")
```

> Creating new collection: 'transactions'

```python
db.transactions.create_index("listing_id")

def process_batch(batch):
    bulk_operations = []
    for listing in batch:
        listing_id = listing['_id']  # Get the _id of the listing
        transactions = listing.get('transactions')  # Get the transactions subdocument

        if transactions:
            transactions['listing_id'] = listing_id
            bulk_operations.append(UpdateOne(
                {"listing_id": listing_id},  # Filter by listing_id
                {"$set": transactions},  # Update or insert the transactions subdocument
                upsert=True))
            db.listings.update_one(
                {"_id": listing_id},
                {"$unset": {"transactions": ""}})
    
    if bulk_operations:
        db.transactions.bulk_write(bulk_operations)

migrate()

print("Transactions have been successfully migrated to the new collection and removed from the original collection.")
```

> Creating new collection: 'hosts'

```python
def process_batch(batch):
    bulk_host_operations = []
    bulk_listing_operations = []
    for listing in batch:
        host_id = listing.get('host_id')
        if host_id:
            host_doc = {
                "_id": host_id,
                **{k: v for k, v in {
                    "about": listing.get('host_about'),
                    "location": listing.get('host_location'),
                    "name": listing.get('host_name'),
                    "has_profile_pic": listing.get('host_has_profile_pic'),
                    "identity_verified": listing.get('host_identity_verified'),
                    "is_superhost": listing.get('host_is_superhost'),
                    "listings_count": listing.get('host_listings_count'),
                    "neighbourhood": listing.get('host_neighbourhood'),
                    "picture_url": listing.get('host_picture_url'),
                    "reponse_rate": listing.get('host_reponse_rate'),
                    "total_listings_count": listing.get('host_total_listings_count'),
                    "url": listing.get('host_url'),
                    "verifications": listing.get('host_verifications'),
                    "response_rate": listing.get('host_response_rate'),
                    "response_time": listing.get('host_response_time'),
                    "thumbnail_url": listing.get('host_thumbnail_url')
                }.items() if v is not None}  # <<< Only update/include non-None fields
            }

            bulk_host_operations.append(UpdateOne(
                {"_id": host_id},
                {"$set": host_doc},
                upsert=True))

            bulk_listing_operations.append(UpdateOne(
                {"_id": listing['_id']},
                {"$unset": {
                    "host_about": "",
                    "host_location": "",
                    "host_name": "",
                    "host_has_profile_pic": "",
                    "host_identity_verified": "",
                    "host_is_superhost": "",
                    "host_listings_count": "",
                    "host_neighbourhood": "",
                    "host_picture_url": "",
                    "host_reponse_rate": "",
                    "host_total_listings_count": "",
                    "host_url": "",
                    "host_verifications": "",
                    "host_response_rate": "",
                    "host_response_time": "",
                    "host_thumbnail_url":""
                    }
                }))
            
    if bulk_host_operations:
        db.hosts.bulk_write(bulk_host_operations, ordered=False)

    if bulk_listing_operations:
        db.listings.bulk_write(bulk_listing_operations, ordered=False)

migrate()

print("Host data has been successfully migrated to the new collection, and host-related fields (except host_id) have been removed from the listings collection.")
```

### Question 3

The company considers investing into property to rent. Which bed type is most common in listings with a waterfront and a dishwasher in New York?

```python
#most common bed type in the listings collection
result = db.listings.aggregate([
    {'$match': {'amenities': {'$all': ['Waterfront', 'Dishwasher']}}},
    {'$group': {
        '_id': '$bed_type',
        'count': {'$sum': 1}
    }},
    {'$project': {
        '_id': 0,
        'bed_type': '$_id',
        'count': 1
    }},
    {'$sort': {'count': -1}},
    {'$limit': 3}
])

for doc in result:
    print(f"In listings with a waterfront and a dishwasher, the most common bed type is '{doc['bed_type']}', with a frequency of {doc['count']} times.")
```

### Question 5

To assess the security of different areas, what is the biggest and smallest (price-security deposit) difference per number of visitors at a property?

```python
biggest_diff_pipeline = [{'$project':{'price':1,
                 'security_deposit':1,
                 'accommodates':1
                 }},
{'$addFields':{'difference':
                   {'$divide':[{'$abs':{'$subtract':['$price', '$security_deposit']}}, '$accommodates']}
                   }},
                   {'$sort':{'difference':-1}},
                   {'$limit':1}]

smallest_diff_pipeline = [{'$match': {
        'security_deposit': {'$ne': None},
    }},
    {'$project':{'price':1,
                 'security_deposit':1,
                 'accommodates':1
                 }},
{'$addFields':{'difference':
                   {'$divide':[{'$abs':{'$subtract':['$price', '$security_deposit']}}, '$accommodates']}
                   }},
                   {'$sort':{'difference':1}},
                   {'$limit':1}]

biggest_diff = [doc.get('difference') for doc in db.listings.aggregate(biggest_diff_pipeline)]
smallest_diff = [doc.get('difference') for doc in db.listings.aggregate(smallest_diff_pipeline)]

print(f"The biggest difference is: {biggest_diff}")
print(f"The smallest difference is: {smallest_diff}")
```
